### Importing the libraries

```python
import tensorflow as tf
tf.__version__
```

```python
import tensorflow.keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')

training_set.class_indices
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('test',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
model2 = tf.keras.applications.DenseNet121(
    weights=None,
    classes=3
)
model2.compile(
  optimizer='adam',
  loss= 'categorical_crossentropy',
  metrics=['accuracy'])
```

```python
model2.summary()
```

```python
model2.fit(x = training_set, validation_data = test_set, epochs = 25)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
from os import listdir
from IPython.display import Image
from IPython.display import display

labels = ["hit", "split", "stay"]

dir = "valid/split/"
filenames = listdir(dir)
for f in filenames:
    test_image = image.load_img(dir + f, target_size = (224, 224))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model2.predict(test_image)
    display(Image(filename=dir + f))
    max_index = np.where(result[0] == max(result[0]))
    print(result[0])
    if max(result[0]) > 0.5:
        print(labels[int(max_index[0])] + " <" + str(max(result[0])) + ">")
    else:
        print("????????????")
```

```python
model2.save('model/hand_model.h5', save_format='h5')
```
